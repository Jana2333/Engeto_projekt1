```python
'''
author = 
'''
TEXTS = ['''
Situated about 10 miles west of Kemmerer, 
Fossil Butte is a ruggedly impressive 
topographic feature that rises sharply 
some 1000 feet above Twin Creek Valley 
to an elevation of more than 7500 feet 
above sea level. The butte is located just 
north of US 30N and the Union Pacific Railroad, 
which traverse the valley. ''',

'''At the base of Fossil Butte are the bright 
red, purple, yellow and gray beds of the Wasatch 
Formation. Eroded portions of these horizontal 
beds slope gradually upward from the valley floor 
and steepen abruptly. Overlying them and extending 
to the top of the butte are the much steeper 
buff-to-white beds of the Green River Formation, 
which are about 300 feet thick.''',

'''The monument contains 8198 acres and protects 
a portion of the largest deposit of freshwater fish 
fossils in the world. The richest fossil fish deposits 
are found in multiple limestone layers, which lie some 
100 feet below the top of the butte. The fossils 
represent several varieties of perch, as well as 
other freshwater genera and herring similar to those 
in modern oceans. Other fish such as paddlefish, 
garpike and stingray are also present.'''
]

USER=input('Zadej jmeno: ').lower()
uzivatele={'bob':'123','ann': 'pass123','mike':'password123','liz':'pass123'}
ODDELOVAC=80*'-'

# overeni uzivatele
if USER in uzivatele.keys():
    PASSWORD=input('Zadej heslo: ').lower()
    print(ODDELOVAC)
else:
  print('Tento uzivatel neexistuje, ukoncuji.')
  exit()

# overeni hesla
if PASSWORD==uzivatele.get(USER):
    print('Vitam te,',USER+'.')
    print('Mame 3 texty k analyze.')
    print(ODDELOVAC)
else:
    print('Chybne heslo, ukoncuji.')
    exit()

vyber=input('Vyber cislo v rozmezi 1 - 3: ')

# overeni ciselneho vstupu
if vyber.isdigit()==False:
    print('To neni cislo, ukoncuji.')
    exit()
else:
    vyber=int(vyber)
    print(ODDELOVAC)

if vyber <1 or vyber > 3:
    print('Spatny vyber, ukoncuji.')
    exit()
else:
    rozdeleni=TEXTS[vyber-1].split()

# cisteni textu
cisteni=[]
for i in rozdeleni:
    cisteni.append(i.strip(",?!:;`.''"))

# pocitani dle zadani
pocet_slov=0
for i in cisteni:
    if len(i)>0: 
        pocet_slov+=1
            
pocet_slov_zacatek_velke=0
for i in cisteni:
    if i.istitle():
        pocet_slov_zacatek_velke+=1
        
bez_cisel=[]
cisla=[]
pocet_cislic=0
for i in cisteni:
    if i.isalpha():
        bez_cisel.append(i)
    elif i.isdigit():
        pocet_cislic+=1
        cisla.append(int(i))
    
pocet_slov_vse_velka=0
pocet_slov_vse_mala=0
for i in bez_cisel:
    if i.isupper() and len(i)>1:
        pocet_slov_vse_velka+=1
    elif i.islower():
        pocet_slov_vse_mala+=1
        
soucet=0
for i in cisla:
    cisla=int(i)
    soucet+=cisla

# zde definuji sklonovani podstatneho jmena 'slovo' a pridavneho jmena 'psany' a casovani slovesa 'byt' do vysledneho textu                
slovo1=''
slovo2=''
slovo3=''
slovo4=''
byt1=''
byt2=''
byt3=''
byt4=''
byt5=''
zacinajici1=''
zacinajici2=''
zacinajici3=''
zacinajici4=''
psana1=''
psana2=''
psana3=''
psana4=''
cislice=''


slova=[slovo1,slovo2,slovo3,slovo4]
byt=[byt1,byt2,byt3,byt4]
zacinajicich=[zacinajici1,zacinajici1,zacinajici3,zacinajici4]
psana=[psana1,psana2,psana3,psana4]
promenne=[pocet_slov,pocet_slov_zacatek_velke,pocet_slov_vse_velka,pocet_slov_vse_mala]

for i in range(len(slova)):
    if promenne[i]>4 or promenne[i]==0:
        slova[i]='slov'
        byt[i]='je'
        zacinajicich[i]='zacinajicich'
        psana[i]='psanych'
    elif promenne[i]>1:
        slova[i]='slova'
        byt[i]='jsou'
        zacinajicich[i]='zacinajici'
        psana[i]='psana'
    else:
        slova[i]='slovo'
        byt[i]='je'
        zacinajicich[i]='zacinajici'
        psana[i]='psane'


if pocet_cislic>4 or pocet_cislic==0:
    cislice='cislic'
    byt5='je'
elif pocet_cislic==1:
    cislice='cislice'
    byt5='je'
else:
    cislice='cislice'
    byt5='jsou'



slovnik={}
delka_slova=[]
for i in cisteni:
    delka_slova.append(len(i))

# serazeni slovniku dle klice
delka_slova.sort() 

for i in delka_slova:
    if i not in slovnik:
        slovnik[i]=1
    else:
        slovnik[i]+=1

                
# priprava pro tisk finalni tabulky
maxklic=max(slovnik.keys())
maxhodnota=max(slovnik.values())
delka_max_klic=len(str(maxklic))
delka_max_hodnota=len(str(maxhodnota))

if len('LEN')>=delka_max_klic and len('OCCURENCE')>=maxhodnota:
    mezera_len=0
    mezera_occ=0
    mezera_occ_1=0
elif len('LEN')>=delka_max_klic and len('OCCURENCE')<maxhodnota:
    mezera_len=0
    rozdil=maxhodnota-len('OCCURENCE')
    if rozdil % 2==0:
        mezera_occ=rozdil//2
        mezera_occ_1=mezera_occ
    else:
        mezera_occ=rozdil//2
        mezera_occ_1=rozdil//2+1
elif len('LEN')<delka_max_klic and len('OCCURENCE')>=maxhodnota:
    mezera_len=delka_max_klic-len('LEN')
    mezera_occ=0
    mezera_occ_1=0
elif len('LEN')<delka_max_klic and len('OCCURENCE')<maxhodnota:
    mezera_len=delka_max_klic-len('LEN')
    rozdil=maxhodnota-len('OCCURENCE')
    if rozdil % 2==0:
        mezera_occ=rozdil//2
        mezera_occ_1=mezera_occ
    else:
        mezera_occ=rozdil//2
        mezera_occ_1=rozdil//2+1
    

delka_oddelovac=len(mezera_len*' '+'LEN'+'|'+mezera_occ*' '+'OCCURENCE'+mezera_occ_1*' '+'|'+'Nr.')

print('Ve vybranen textu',byt[0],pocet_slov,str(slova[0])+'.')
print('Ve vybranem textu',byt[1],pocet_slov_zacatek_velke,slova[1],zacinajicich[1],'velkym pismenem.')
print('Ve vybranem textu',byt[2],pocet_slov_vse_velka,slova[2],psana[2],'velkymi pismeny.')
print('Ve vybranem textu',byt[3],pocet_slov_vse_mala,slova[3],psana[3],'malymi pismeny.')
print('Ve vybranem textu',byt5,pocet_cislic,cislice+'.')
print('Soucet vsech cisel je',str(soucet)+'.')
print(ODDELOVAC)
print(mezera_len*' '+'LEN'+'|'+mezera_occ*' '+'OCCURENCE'+mezera_occ_1*' '+'|'+'Nr.')
print(delka_oddelovac*'-')

for i, j in slovnik.items():
    if len('LEN')>=delka_max_klic: 
        mezera_i=len('LEN')-len(str(i))
        if len('OCCURENCE')>=maxhodnota:
            mezera_j=len('OCCURENCE')-j
        else:
            mezera_j=maxhodnota-j
    else: 
        mezera_i=delka_max_klic-len(str(i))
        if len('OCCURENCE')>=maxhodnota:
            mezera_j=len('OCCURENCE')-j
        else:
            mezera_j=maxhodnota-j
    print(mezera_i*' '+str(i)+'|'+j*'*'+mezera_j*' '+'|'+str(j))


    





```
